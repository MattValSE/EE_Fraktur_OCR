```python
import os
from openai import AzureOpenAI
from dotenv import load_dotenv


load_dotenv()

# Retrieve API key
GPT_API_KEY = os.getenv("GPT")
if not GPT_API_KEY:
    raise ValueError("Environment variable 'GPT' is not set or invalid.")

endpoint = "https://tu-openai-ltat-tartunlp.openai.azure.com/"
#endpoint = "https://tu-openai-lmm-ltat-tartunlp.openai.azure.com"
model_name = "gpt-4o"
deployment = "gpt-4o-mattias"

subscription_key = GPT_API_KEY
api_version = "2025-04-01-preview"

client = AzureOpenAI(
    api_version=api_version,
    azure_endpoint=endpoint,
    api_key=subscription_key,
)

response = client.chat.completions.create(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "I am going to Paris, what should I see?",
        }
    ],
    max_tokens=4096,
    temperature=1.0,
    top_p=1.0,
    model=deployment
)

print(response.choices[0].message.content)
```
